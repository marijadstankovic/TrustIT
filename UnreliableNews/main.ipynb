```python
import BabylonBee
import DailySquib
import Foxnews
import NewYorkerBorowitzReport
import Telegraph
import TheBeaverton
import TheChaser
import TheHardTimes
import TheNewRepublic
import TheRochdaleHerald
import TheSun
import WeeklyWorldNews

import ReadWriteCSV
```

```python

def daily_scraping():
    array_of_ids = ReadWriteCSV.read()
    print(len(array_of_ids))
    if(len(array_of_ids)<10):
        print(array_of_ids)
    try:
        print("BabylonBee")
        BabylonBee.scraping(array_of_ids)
    except:
        print("BabylonBee wrong")

    try:
        print("DailySquib")
        DailySquib.scraping(array_of_ids)
    except:
        print("DailySquib wrong")

    try:
        print("Foxnews")
        Foxnews.scraping(array_of_ids)
    except:
        print("Foxnews wrong")

    try:
        print("NewYorkerBorowitzReport")
        NewYorkerBorowitzReport.scraping(array_of_ids)
    except:
        print("NewYorkerBorowitzReport wrong")

    try:
        print("Telegraph")
        Telegraph.scraping(array_of_ids)
    except:
        print("Telegraph wrong")

    try:
        print("TheBeaverton")
        TheBeaverton.scraping(array_of_ids)
    except:
        print("TheBeaverton wrong")

    try:
        print("TheChaser")
        TheChaser.scraping(array_of_ids)
    except:
        print("TheChaser wrong")

    try:
        print("TheHardTimes")
        TheHardTimes.scraping(array_of_ids)
    except:
        print("TheHardTimes wrong")

    try:
        print("TheNewRepublic")
        TheNewRepublic.scraping(array_of_ids)
    except:
        print("TheNewRepublic wrong")

    try:
        print("TheRochdaleHerald")
        TheRochdaleHerald.scraping(array_of_ids)
    except:
        print("TheRochdaleHerald wrong")

    try:
        print("TheSun")
        TheSun.scraping(array_of_ids)
    except:
        print("TheSun wrong")

    try:
        print("WeeklyWorldNews")
        WeeklyWorldNews.scraping(array_of_ids)
    except:
        print("WeeklyWorldNews wrong")
        print(len(array_of_ids))
    print(len(array_of_ids))
    ReadWriteCSV.dump()
```

```python
daily_scraping()
```

```python
import schedule
import time

schedule.every(8).hours.do(daily_scraping)

while 1:
    schedule.run_pending()
    time.sleep(1)
```
