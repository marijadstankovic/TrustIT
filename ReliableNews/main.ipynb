```python
import AlJazeera
import Vox
#import AssociatedPress###
import TheHill
import TheAtlantic
import Snopes
import Reuters
import NewYorkTimes
import Economist
import DesertNews
import Deadline
import Bellingcat
import BBC

import RReadWriteCSV
```

```python
def daily_scraping():
    array_of_ids = RReadWriteCSV.read()
    print(len(array_of_ids))
    if(len(array_of_ids)<10):
        print(array_of_ids)

    try:
        print("Vox")
        Vox.scraping(array_of_ids)
    except:
        print("Vox wrong")
    # try:
    #     print("AssociatedPress")
    #     AssociatedPress.scraping(array_of_ids)
    # except:
    #     print("AssociatedPress wrong")
    try:
        print("TheHill")
        TheHill.scraping(array_of_ids)
    except:
        print("TheHill wrong")
    try:
        print("TheAtlantic")
        TheAtlantic.scraping(array_of_ids)
    except:
        print("TheAtlantic wrong")
    try:
        print("Snopes")
        Snopes.scraping(array_of_ids)
    except:
        print("Snopes wrong")
    try:
        print("Reuters")
        Reuters.scraping(array_of_ids)
    except:
        print("Reuters wrong")
    try:
        print("NewYorkTimes")
        NewYorkTimes.scraping(array_of_ids)
    except:
        print("NewYorkTimes wrong")
    try:
        print("Economist")
        Economist.scraping(array_of_ids)
    except:
        print("Economist wrong")
    try:
        print("DesertNews")
        DesertNews.scraping(array_of_ids)
    except:
        print("DesertNews wrong")
    try:
        print("Deadline")
        Deadline.scraping(array_of_ids)
    except:
        print("Deadline wrong")
    try:
        print("Bellingcat")
        Bellingcat.scraping(array_of_ids)
    except:
        print("Bellingcat wrong")
    try:
        print("BBC")
        BBC.scraping(array_of_ids)
    except:
        print("BBC wrong")
    print(len(array_of_ids))
    RReadWriteCSV.dump()

    
```

```python
daily_scraping()
```

```python
    
import schedule
import time

schedule.every(8).hours.do(daily_scraping)

while 1:
    schedule.run_pending()
    time.sleep(1)
```
